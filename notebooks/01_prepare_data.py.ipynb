```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D


# Define the two objective functions
def objective_function_1(x, y):
    return (x - 2) ** 2 + (y - 3) ** 2


def objective_function_2(x, y):
    return np.sin(x) * np.cos(y) + 0.1 * (x ** 2 + y ** 2)


# Generate grid
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z1 = objective_function_1(X, Y)
Z2 = objective_function_2(X, Y)

# Plotting the objectives
fig = plt.figure(figsize=(15, 5))

# Objective 1
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='coolwarm', edgecolor='none')
ax1.set_title("Objective Function 1: Minimize Distance from (2, 3)")
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f1(x, y)')

# Objective 2
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='viridis', edgecolor='none')
ax2.set_title("Objective Function 2: Sinusoidal Landscape")
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('f2(x, y)')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample points
x_samples = np.random.uniform(-5, 5, 100)
y_samples = np.random.uniform(-5, 5, 100)
f1_samples = objective_function_1(x_samples, y_samples)
f2_samples = objective_function_2(x_samples, y_samples)

# Plot Pareto front
plt.figure(figsize=(8, 6))
plt.scatter(f1_samples, f2_samples, c='blue', label="Sampled Points")
plt.xlabel("Objective 1: f1(x, y)")
plt.ylabel("Objective 2: f2(x, y)")
plt.title("Pareto Front for Multi-Objective Optimization")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Définition des deux fonctions objectives
def objective_function_1(x, y):
    return (x - 2) ** 2 + (y - 3) ** 2


def objective_function_2(x, y):
    return np.sin(x) * np.cos(y) + 0.1 * (x ** 2 + y ** 2)


# Génération de 2000 points aléatoires dans le domaine x, y ∈ [-5, 5]
num_points = 2000
x_samples = np.random.uniform(0, 5, num_points)
y_samples = np.random.uniform(0, 5, num_points)

# Calcul des valeurs pour chaque fonction
f1_values = objective_function_1(x_samples, y_samples)
f2_values = objective_function_2(x_samples, y_samples)

# Visualisation des points générés dans l'espace des objectifs
plt.figure(figsize=(8, 6))
plt.scatter(f1_values, f2_values, c='blue', s=10, alpha=0.6, label="Points échantillonnés")
plt.xlabel("Objectif 1: f1(x, y)")
plt.ylabel("Objectif 2: f2(x, y)")
plt.title("Front de Pareto pour l'optimisation multi-objectifs")
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({'x': x_samples, 'y': y_samples, 'f1': f1_values, 'f2': f2_values})
data.to_parquet("../data/data.parquet")
```
